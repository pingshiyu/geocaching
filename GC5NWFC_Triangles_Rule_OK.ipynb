```python
import matplotlib.pyplot as plt
import numpy as np
import scipy
from typing import Optional
from math import acos, pi
```

```python
# to plot:
# arcccos((1-b)/(2b)) + arccos((b+1)/(2a)) + arccos((a+1)/(2b)) - pi = 0
# a >= 2b - 1

def in_range(cosX) -> bool:
    return -1 <= cosX <= 1

def triangle_fn(a: float, b: float) -> float:
    cosA = (1-b) / (2*b) # (1-5) / (2*5) = -4 / 10
    cosB = (b+1) / (2*a) # (5+1) / (2*5) = 6 / 10
    cosC = (a+1) / (2*b) # (5+1) / (2*5) = 6 / 10
    
    if in_range(cosA) and in_range(cosB) and in_range(cosC):
        val = np.arccos(cosA) + np.arccos(cosB) + np.arccos(cosC) - pi
    else:
        val = 10
    
    # print(a, b, cosA, cosB, cosC, val)
    return val
```

```python
# _b as a function of a
def _b(a: float) -> Optional[float]:
    # default guess: b = a should be safe to ensure all stays in range
    res = scipy.optimize.root(
        lambda b: triangle_fn(a, b), 
        max(a - 0.5, 1))
        #options={'return_all': False})
    # print(res)
    if (res.success) and (res.fun < 1e-2):
        return res.x[0]
    else:
        return np.nan
```

```python
a_vals = np.linspace(1, 10, num=100)
b_vals = np.array(list(map(_b, a_vals)))
np.stack((a_vals, b_vals)).T
```
